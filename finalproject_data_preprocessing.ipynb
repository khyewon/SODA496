```python
import pandas as pd
```

```python
# 
execution = pd.read_csv("execution.csv")
execution
```

```python
Execution = pd.melt(execution, id_vars=['country', 'region', 'population', 'latest execution'],
                    value_vars = ['2022', '2021', '2020' ,'2019', '2018', '2017', '2016' ,'2015', '2014', '2013', '2012', '2011', '2010', '2009', '2008', '2007'],
                    var_name = 'year', value_name = 'executed')
Execution = Execution.sort_values(by = ['country', 'year'])
Execution
```

```python
Execution.to_csv('execution_clean.csv', index = False)
```

```python
literacy = pd.read_csv("literacyrate.csv")
literacy
```

```python
Literacy = pd.melt(literacy, id_vars=['Country Name'],
                    value_vars = ['2021', '2020' ,'2019', '2018', '2017', '2016' ,'2015', '2014', '2013', '2012', '2011', '2010', '2009', '2008', '2007'],
                    var_name = 'year', value_name = 'literacy rate')
Literacy.columns = ['country', 'year', 'literacy rate']
Literacy = Literacy.sort_values(by = ['country', 'year'])
Literacy
```

```python
Literacy.to_csv('literacy_clean.csv', index = False)
```
